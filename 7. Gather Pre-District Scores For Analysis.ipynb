### Gather the individual singles and doubles matches from the downloaded JSON files

#### First, gather the singles results

### Remove records from the singles/doubles match files if they are from a district or state tournament

```python
import pandas as pd

MAIN_DIR = r"C:\Users\toddw\Desktop\Python Rating Code and Files\GenAI_Rewrite"

# Input files
SINGLES_IN  = MAIN_DIR + r"\singles_matches_clean.csv"
DOUBLES_IN  = MAIN_DIR + r"\doubles_matches_clean.csv"
TOURNEY_IN  = MAIN_DIR + r"\district_state_tournament_match_ids.csv"

# Output files
SINGLES_OUT = MAIN_DIR + r"\Non_Tourney_Singles.csv"
DOUBLES_OUT = MAIN_DIR + r"\Non_Tourney_Doubles.csv"

# Load tournament match IDs
tourney_ids = pd.read_csv(TOURNEY_IN)

# Make sure the ID field is numeric and standardized
tourney_ids["team_match_id"] = pd.to_numeric(tourney_ids["team_match_id"], errors="coerce")

# Convert to a set for fast lookup
tourney_set = set(tourney_ids["team_match_id"].dropna().astype(int))


# -------------------------------------------------------
# Process SINGLES
# -------------------------------------------------------
sing = pd.read_csv(SINGLES_IN)

# Ensure numeric
sing["team_match_id"] = pd.to_numeric(sing["team_match_id"], errors="coerce")

# Filter out matches whose team_match_id is in tournament list
sing_filtered = sing[~sing["team_match_id"].isin(tourney_set)]

# Write output
sing_filtered.to_csv(SINGLES_OUT, index=False)


# -------------------------------------------------------
# Process DOUBLES
# -------------------------------------------------------
doub = pd.read_csv(DOUBLES_IN)

# Ensure numeric
doub["team_match_id"] = pd.to_numeric(doub["team_match_id"], errors="coerce")

# Filter out tournament matches
doub_filtered = doub[~doub["team_match_id"].isin(tourney_set)]

# Write output
doub_filtered.to_csv(DOUBLES_OUT, index=False)


print("Filtering complete.")
print(f"Non-tournament singles written to: {SINGLES_OUT}")
print(f"Non-tournament doubles written to: {DOUBLES_OUT}")
```
